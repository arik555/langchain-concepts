```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import InMemoryChatMessageHistory 

load_dotenv()

# connect to the remote 
llm = HuggingFaceEndpoint(
    # thinking enabled by default. -- expensive and time consuming.
    # repo_id="Qwen/Qwen3-30B-A3B",
    
    repo_id="Qwen/Qwen2.5-7B-Instruct",  # fast, no thinking mode -- closer to gpt-3.5-turbo
    task="text-generation",
    temperature=0
)

# initiate the chatmodel
chat_model = ChatHuggingFace(
    llm=llm,
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful health firness guide and expert."),
    MessagesPlaceholder(variable_name="history_user_messages"),
    ("human", "{input_user_message}")
])

parser = StrOutputParser()
```

```python
chain = prompt | chat_model | parser
```

```python
store = {}

def get_session_messages(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
```

```python
chat_with_messages_history = RunnableWithMessageHistory(
    chain,
    get_session_messages,
    input_messages_key="input_user_message",
    history_messages_key="history_user_messages"
)
```

```python
response = chat_with_messages_history.invoke(input={
    "input_user_message": "what is BMI index?"
}, config={"configurable": {"session_id": "0001"}})
```
