```python
from langchain_ollama import ChatOllama
from langchain.tools import tool
from langchain_core.messages import SystemMessage
from langchain.agents import create_agent
from dotenv import load_dotenv

load_dotenv()
```

```python
llm = ChatOllama(model="llama3.1:latest", temperature=0)
```

```python
@tool
def add_numbers(a: int | float, b: int | float):
    """ Function to return the sum of two numbers"""
    return a + b

@tool
def subtract_numbers(a: int | float, b: int | float):
    """ Function to return the subtraction of two numbers"""
    return a - b

@tool
def multiple_numbers(a: int | float, b: int | float):
    """ Function to return the product of two numbers"""
    return a * b

@tool
def divide_numbers(a: int | float, b: int | float):
    """ Function to return the division between two numbers. Returns zero if the divisor is 0
    Args:
        a: dividend
        b: divisor
    """
    return a / b
```

```python
prompt = SystemMessage(content="You are a helpful math teacher. Use your tools to perform calculations and solve problems.")

agent = create_agent(
    model=llm,
    tools=[add_numbers, subtract_numbers, multiple_numbers, divide_numbers],
    system_prompt=prompt,
    debug=True
)
```

```python
user_query = {'messages': [{'role': 'user', 'content': 'How much is 2 + 30?'}]}
resp = agent.invoke(input=user_query)
```

```python
resp["messages"][-1].content
```
